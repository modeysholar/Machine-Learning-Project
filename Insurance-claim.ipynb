**DATA PREPROCESSING**



Target encoding 



```python

# Replacing Insured occupation with it's mean target value.
occupation_map = train_data.groupby(['insured_occupation']).\
    agg({'total_claim_amount' : 'mean'}).to_dict()['total_claim_amount']
train_data = train_data.copy()
train_data['insured_occupation'] = train_data['insured_occupation'].map(occupation_map)

# Replacing Auto model with it's mean target value.
model_map = train_data.groupby(['auto_model']).\
    agg({'total_claim_amount' : 'mean'}).to_dict()['total_claim_amount']
train_data = train_data.copy()
train_data['auto_model'] = train_data['auto_model'].map(model_map)

# Replacing Auto make with it's mean target value.
make_map = train_data.groupby(['auto_make']).\
    agg({'total_claim_amount' : 'mean'}).to_dict()['total_claim_amount']
train_data = train_data.copy()
train_data['auto_make'] = train_data['auto_make'].map(make_map)

# Replacing Policy state with it's mean target value.
policy_state_map = train_data.groupby(['policy_state']).\
    agg({'total_claim_amount' : 'mean'}).to_dict()['total_claim_amount']
train_data = train_data.copy()
train_data['policy_state'] = train_data['policy_state'].map(policy_state_map)

# Replacing insured hobbies with it's mean target value.
hobbies = train_data.groupby(['insured_hobbies']).\
    agg({'total_claim_amount' : 'mean'}).to_dict()['total_claim_amount']
train_data = train_data.copy()
train_data['insured_hobbies'] = train_data['insured_hobbies'].map(hobbies)

# Replacing Policy number with it's mean target value.
policy_number = train_data.groupby(['policy_number']).\
    agg({'total_claim_amount' : 'mean'}).to_dict()['total_claim_amount']
train_data = train_data.copy()
train_data['policy_number'] = train_data['policy_number'].map(policy_number)

# Replacing insured sex with it's mean target value.
sex = train_data.groupby(['insured_sex']).\
    agg({'total_claim_amount' : 'mean'}).to_dict()['total_claim_amount']
train_data = train_data.copy()
train_data['insured_sex'] = train_data['insured_sex'].map(sex)

# Replacing incident location with it's mean target value.
location = train_data.groupby(['incident_location']).\
    agg({'total_claim_amount' : 'mean'}).to_dict()['total_claim_amount']
train_data = train_data.copy()
train_data['incident_location'] = train_data['incident_location'].map(location)

# Replacing Authorities contacted with it's mean target value.
authority = train_data.groupby(['authorities_contacted']).\
    agg({'total_claim_amount' : 'mean'}).to_dict()['total_claim_amount']
train_data = train_data.copy()
train_data['authorities_contacted'] = train_data['authorities_contacted'].map(authority)

# Replacing Incident state with it's mean target value.
state = train_data.groupby(['incident_state']).\
    agg({'total_claim_amount' : 'mean'}).to_dict()['total_claim_amount']
train_data = train_data.copy()
train_data['incident_state'] = train_data['incident_state'].map(state)

# Replacing Incident city with it's mean target value.
city = train_data.groupby(['incident_city']).\
    agg({'total_claim_amount' : 'mean'}).to_dict()['total_claim_amount']
train_data = train_data.copy()
train_data['incident_city'] = train_data['incident_city'].map(city)

# Replacing Insured relationship with it's mean target value.
relationship = train_data.groupby(['insured_relationship']).\
    agg({'total_claim_amount' : 'mean'}).to_dict()['total_claim_amount']
train_data = train_data.copy()
train_data['insured_relationship'] = train_data['insured_relationship'].map(relationship)

# Replacing witnesses with it's mean target value.
witnesses = train_data.groupby(['witnesses']).\
    agg({'total_claim_amount' : 'mean'}).to_dict()['total_claim_amount']
train_data = train_data.copy()
train_data['witnesses'] = train_data['witnesses'].map(witnesses)

train_data.head()
```

Label Encoding ordinal categorical coulmns



```python
# Encoding education_level col
education = {'High School': 0, 'College': 1, 'JD': 2, 'Masters':3, 'MD': 4, 'PhD': 5, 'Associate': 6}
train_data['insured_education_level'] = train_data['insured_education_level'].map(education)

# Encoding property damage col
Damage = {'NO': 0, 'YES': 1}
train_data['property_damage'] = train_data['property_damage'].map(Damage)

#Encoding police_report col
report = {'NO': 0, 'YES': 1}
train_data['police_report_available'] = train_data['police_report_available'].map(report)

#Encoding collision_type col
collision = {'Rear Collision': 0, 'Side Collision': 1, 'Front Collision': 2}
train_data['collision_type'] = train_data['collision_type'].map(collision)

#Encoding incident_severity col
severity = {'Trivial Damage': 0, 'Minor Damage': 1, 'Total Loss': 2, 'Major Damage': 3}
train_data['incident_severity'] = train_data['incident_severity'].map(severity)

#Encoding incident_type col
incident = {'Parked Car': 0, 'Vehicle Theft': 1, 'Multi-vehicle Collision': 2, 'Single Vehicle Collision': 3}
train_data['incident_type'] = train_data['incident_type'].map(incident)

#Encoding policy_csl col
csl = {'100/300': 0, '250/500': 1, '500/1000': 2}
train_data['policy_csl'] = train_data['policy_csl'].map(csl)
```

```python
#Calculating the number of months the policy started before the incident occured 
import datetime 

train_data['incident_date'] = pd.to_datetime(train_data['incident_date'], format='%Y-%m-%d')
train_data['policy_bind_date'] = pd.to_datetime(train_data['policy_bind_date'], format='%Y-%m-%d')

train_data['policy_month_before_incident'] = train_data['incident_date'] - train_data['policy_bind_date']
train_data['policy_month_before_incident']=train_data['policy_month_before_incident']/np.timedelta64(1,'M')

#Dropping the policy bind date and incident date and replaced with a new column which is the difference 
train_data = train_data.drop(['incident_date', 'policy_bind_date'], axis=1)  

train_data.head()
```

```python
X = train_data.drop(['total_claim_amount', 'Customer_ID', 'policy_number'], axis=1)  
y = train_data['total_claim_amount']  

corrmat = train_data.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(25,25))
#plot heat map
g=sns.heatmap(train_data[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

Feature Selection



```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
selector = SelectKBest(f_regression, k = 20)

X_news = selector.fit_transform(X, y)
names = X.columns.values[selector.get_support()]
scores = selector.scores_[selector.get_support()]
names_scores = list(zip(names, scores))
ns_df = pd.DataFrame(data = names_scores, columns=['Feat_names', 'F_Scores'])

#Sort the dataframe for better visualization
ns_df_sorted = ns_df.sort_values(['F_Scores', 'Feat_names'], ascending = [False, True])
print(ns_df_sorted)
```

**DATA MODELLING**

```python
# Splitting Data into Train and Test

from sklearn.model_selection import train_test_split
X_tr, X_test, y_tr, y_test = train_test_split(X_new, y, test_size=0.3, random_state= 40)
```

FEATURE SCALING


```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()

X_tr= sc.fit_transform(X_tr)
X_test= sc.transform(X_test)
```

LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(X_tr, y_tr)
```

```python
y_pred1= regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

print(mean_absolute_error(y_test,y_pred1))
print(mean_squared_error(y_test,y_pred1))
print(np.sqrt(mean_squared_error(y_test,y_pred1)))
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = regressor, X = X_tr, y = y_tr, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
